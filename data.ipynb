```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")

import numpy as np
import pandas as pd

wins_1 = pd.read_csv('wins_1.csv')
wins_10 = pd.read_csv('wins_10.csv')
wins_20 = pd.read_csv('wins_20.csv')
wins_1["ws"] = "1_win_game"
wins_10["ws"] = "10_win_game"
wins_20["ws"] = "20_win"
data = pd.concat([wins_1, wins_10])
data = data.sort_values(by="wins", ignore_index=True)
data
```
